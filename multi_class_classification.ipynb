```python
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
```

```python
df = pd.read_csv('https://drive.google.com/uc?id=1roJ83AbgzDcvRr0Gwud0BmdUQx-oSG-w')
df
```

```python
# menghapus kolom yang tidak digunakan
df = df.drop(columns='Id')
df.head()
```

```python
# mengubah data kategorikal menggunakan teknik one hot encoding
category = pd.get_dummies(df.Species, dtype=int)
category
```

```python
# menggabungkan kolom hasil one hot encoding dengan kolom yang sudah ada dan membuang kolom species

new_df = pd.concat([df, category], axis=1)
new_df = new_df.drop(columns='Species')
new_df
```

```python
# konversi dataframe ke numpy array
dataset = new_df.values
dataset
```

```python
# memisahkan attribute dan label
# pilih 4 kolom pertama untuk dijadikan attribute
X = dataset[:, 0:4]

# pilih 3 kolom terakhir untuk dijadikan label
y = dataset[:, 4:7]
```

```python
# normalisasi
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
X_scale
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_scale, y, test_size=0.3)
```

```python
model = Sequential([
    Dense(64, activation='relu', input_shape=(4,)),
    Dense(64, activation='relu'),
    Dense(3, activation='softmax')
])
```

```python
model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train, Y_train, epochs=100)
```

```python
model.evaluate(X_test, Y_test, batch_size=1)
```
